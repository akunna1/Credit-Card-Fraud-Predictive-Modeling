### Building a model to detect credit card fraud using machine learning

#### Data Source

```python
# Kaggle link: https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud
```

#### Importing libraries

```python
import numpy as np # for numerical operations
import pandas as pd # for data manipulation
import matplotlib.pyplot as plt # for visualization
import seaborn as sns # for visualization

# scikit-learn (sklearn) for machine learning tasks
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas Dataframe
credit_card_data = pd.read_csv("creditcard.csv")
```

#### About dataset

```python
# Features V1, V2, … V28 are the principal components obtained with PCA, the only features which have not 
# been transformed with PCA are 'Time' and 'Amount'. The time is in seconds.
```

```python
# There is  a column named "Class" indicating whether a transaction is fraudulent (1) or legitimate (0).
```

```python
# distribution of legit transactions & fraudulent transactions
credit_card_data['Class'].value_counts()
```

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# comparing the values for both transactions using groupby
credit_card_data.groupby('Class').mean()
```

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating the sampled legitimate transactions with the fraudulent transactions to balance the dataset for training 
# the model
new_dataset = pd.concat([legit_sample, fraud], axis=0)
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X = new_dataset.drop(columns='Class', axis=1) # Features (independent variables)
Y = new_dataset['Class'] # Target variable (dependent variable)
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

#### Model Training (Logistic Regression)

```python
# Creating a logistic regression model instance
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
                   intercept_scaling=1, l1_ratio=None, max_iter=100,
                   multi_class='auto', n_jobs=None, penalty='l2',
                   random_state=None, solver='lbfgs', tol=0.0001, verbose=0,
                   warm_start=False)
```

#### Model Evaluation

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
